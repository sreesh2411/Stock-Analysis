## <u>Moving Average</u>

A moving average (MA) is a widely used technical indicator that smooths out price trends by filtering out the noise from random short-term price fluctuations and helps cut down the amount of "noise" on a price chart. 

```python
#Vizualing the noise in the stock prices.

gm['Open'].iloc[440:500].plot(figsize = (15,7))
plt.grid()
```

  As we can see, there is a lot of noise in the chart, this noise can be smoothened with MA.

```python
#Calculate moving average over 50 and 200 days using rolling() function

gm['Open'].plot(figsize = (15,7), label = "Opening Price")
gm['MA50'] = gm['Open'].rolling(50).mean()
gm['MA50'].plot(label = 'MA50')
gm['MA200'] = gm['Open'].rolling(200).mean()
gm['MA200'].plot(label = 'MA200')
plt.legend()
plt.grid()
```

 The graph has smoothened a bit over a rolling window of 50 and even more over a rolling window of 200. However, too much smoothening can cause a loss in detection of trends.

### Candlestick Chart
 A <u>candlestick chart</u> (also called Japanese candlestick chart) is a style of financial chart used to describe price movements of a security, derivative, or currency. Each "candlestick" typically shows one day. Traders use the candlesticks to make trading decisions based on regularly occurring patterns that help forecast the short-term direction of the price. A one-month chart may show the 20 trading days as 20 candlesticks.

```python
#Plotting a candlestick chart.

from mpl_finance import candlestick_ohlc
from matplotlib.dates import DateFormatter, date2num, WeekdayLocator, DayLocator, MONDAY
```

```python
# We see the candlestick chart for ford prices in January 2012.

ford_reset = ford.loc['2012-01' : '2012-01'].reset_index()

ford_reset['date_ax'] = ford_reset['Date'].apply(lambda date : date2num(date))
ford_values = [tuple(vals) for vals in ford_reset[['date_ax', 'Open', 'High', 'Low', 'Close']].values]

mondays = WeekdayLocator(MONDAY)
alldays = DayLocator()
weekFormatter = DateFormatter('%b %d')
dayFormatter = DateFormatter('%d')

fig, ax = plt.subplots()
candlestick_ohlc(ax, ford_values, width = 0.6, colorup = 'g', colordown = 'r')
```

## Daily Percentage Change

Percentage change is a simple mathematical concept that represents the degree of change over time. It is used for many purposes in finance, often to represent the price change of a security. We calculate the daily percentage change is using:
change = (present day price)/(previous day price) -1
$ r_{t} =  \frac{p_{t}}{p_{t-1}}  - 1 $ 


```python
#shift() function shifts the row by 1 to the previous row.

tesla['returns'] = (tesla['Close'] / tesla['Close'].shift(1))- 1
```

```python
tesla.head()
```

```python
ford['returns'] = (ford['Close'] / ford['Close'].shift(1))- 1
gm['returns'] = (gm['Close'] / gm['Close'].shift(1))- 1
```

```python
ford.head()
```

```python
gm.head()
```

## Volatility

In finance, volatility (usually denoted by ) is the degree of variation of a trading price series over time, usually measured by the standard deviation of logarithmic returns.

```python
ford['returns'].hist(bins = 50)
```

```python
gm['returns'].hist(bins = 50)
```

```python
tesla['returns'].hist(bins = 50)
```

 Let us try to plot all the three graphs together so as to compare them.

```python
tesla['returns'].hist(bins = 100, label = 'Tesla', figsize = (15,7), alpha = 0.5)
ford['returns'].hist(bins = 100, label = 'Ford', alpha = 0.5)
gm['returns'].hist(bins = 100, label = 'GM', alpha = 0.5)
plt.legend()
```

 We can see that Tesla's graph is wider than Ford and GM and hence, more volatile. Let us visualize this in a normal distribution manner.

```python
tesla['returns'].plot(kind='kde', label='Tesla', figsize=(15,7))
ford['returns'].plot(kind='kde', label='Ford')
gm['returns'].plot(kind='kde', label='GM')

plt.legend()
```

The curves of Ford and GM are steeper and higher and Tesla's is lower and wider. Thus, this supports our point that Tesla's returns are more volatile.

# Cumulative Return
The cumulative return is the total change in the investment price over a set timean aggregate return, not an annualized or a daily one. We can find out how much a dollar that we invested in a company would be worth during the 5 years. Let us understand this with an example.
$ i_{t} = (1 + r_{t})i_{t-1} = (1 + \frac{p_{t}}{p_{t-1}} -1)i_{t-1} = \frac{p_{t}}{p_{t-1}}i_{t-1} $

| Date | Price($) |                 
| --- | --- |
| 1-jan-2012 | 10 |
| 2-jan-2012 | 15 |
| 3-jan-2012 | 20 |
| 4-jan-2012 | 25 |

| Date | Daily Return | %Daily return |
| --- | --- | -- |
| 1-jan-2012 | 10 | - |
| 2-jan-2012 | 15/10 = 1.5 | 50 |
| 3-jan-2012 | 20/15 = 1.33 | 33 |
| 4-jan-2012 | 25/20 = 1.25 | 25 |

| Date | Cumulative Return | %Cumulative return |
| --- | --- | -- |
| 1-jan-2012 | 10 | - |
| 2-jan-2012 | 15/10 = 1.5 | 150 |
| 3-jan-2012 | 20/10 = 2.0 | 200 |
| 4-jan-2012 | 25/10 = 2.5 | 250 |




```python
tesla['Cumulative Return'] = (1 + tesla['returns']).cumprod()
```

```python
tesla.head()
```

```python
ford['Cumulative Return'] = (1 + ford['returns']).cumprod()
gm['Cumulative Return'] = (1 + gm['returns']).cumprod()
```

```python
tesla['Cumulative Return'].plot(label = 'Tesla', figsize = (15,7))
ford['Cumulative Return'].plot(label = 'Ford')
gm['Cumulative Return'].plot(label = 'GM')   
plt.legend()
plt.ylabel('Cumulative Return')
plt.title('Cumulative Return over Time')
```

 Tesla returns are much greater than Ford and GM returns.
